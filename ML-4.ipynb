```python
def function_to_minimize(x):
    return (x+3)**2

def gradient(x):
    return 2*(x+3)

x_values = []
y_values = []

x=2
learning_rate = 0.01
num_iterations = 1000

for i in range(num_iterations):
    grad = gradient(x)
    x = x - (learning_rate*grad)
    y = function_to_minimize(x)
    x_values.append(x)
    y_values.append(y)
    print(f"Iteration {i+1}: x = {x:.4f}, y = {y:.4f}")
print(f"Local minimum found at x = {x:.4f}, y = {y:.4f}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 20))
plt.plot(range(num_iterations), y_values, 'b.-')
plt.axvline(x=len(x_values)-1, color='r', linestyle='-', label='Local Minimum')
plt.xlabel('Iteration')
plt.ylabel('Function Value')
plt.title('Gradient Descent Optimization')
plt.grid(True)
plt.show()
```
